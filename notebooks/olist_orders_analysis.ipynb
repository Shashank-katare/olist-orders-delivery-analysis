```python
delivered_orders = orders[orders["order_status"] == "delivered"]
```

```python
delivered_orders["order_status"].value_counts()
```

```python
date_cols = [
    "order_purchase_timestamp",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_cols:
    delivered_orders[col] = pd.to_datetime(delivered_orders[col])
```

```python
delivered_orders[date_cols].isnull().sum()
```

```python
delivered_orders = delivered_orders.dropna(
    subset=["order_delivered_customer_date"]
)
```

```python
delivered_orders["delivery_delay_days"] = (
    delivered_orders["order_delivered_customer_date"]
    - delivered_orders["order_estimated_delivery_date"]
).dt.days
```

```python
delivered_orders["delivery_delay_days"].describe()
```

```python
delivered_orders["delivery_status"]=delivered_orders["delivery_delay_days"].apply(lambda x: "Late" if x > 0 else "On-Time/Early")
```

```python
delivered_orders["delivery_status"].value_counts()
```

```python
plt.Figure(figsize=(6,4))
sns.countplot(
    data = delivered_orders,
    x = "delivery_status"
)
plt.title("Late vs On-Time/Early Deliveries")
plt.xlabel("delivery_status")
plt.ylabel("Number of orders")
plt.tight_layout()
plt.show()
```

```python
delivered_orders["delivery_delay_days"].describe()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(delivered_orders["delivery_delay_days"], bins=50, kde=True)
plt.axvline(0, color="red", linestyle="--", label="On-time threshold")
plt.xlabel("Delivery Delay (Days)")
plt.ylabel("Number of Orders")
plt.title("Distribution of Delivery Delays")
plt.legend()
plt.show()
```
